```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Using forward slashes (recommended)
df = pd.read_csv('../dataset/data.csv')
```

```python
df.head()
```

```python
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist()[:20], "...")
print("\nData Types:\n", df.dtypes.value_counts())
print("\nFirst 5 rows:\n", df.head())
print("\nMissing Values:\n", df.isnull().sum().sort_values(ascending=False).head(10))
```

```python
df.describe()
```

```python
plt.figure(figsize=(12,8))
numeric_df = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_df.corr().iloc[:20, :20], cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap (Top 20 Features)")
plt.show()

# --- Feature Distributions ---
num_cols = df.select_dtypes(include='number').columns[:5]
df[num_cols].hist(bins=30, figsize=(12,8))
plt.suptitle("Feature Distributions")
plt.show()
```
